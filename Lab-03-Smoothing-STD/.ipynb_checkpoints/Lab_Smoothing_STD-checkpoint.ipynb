```python
image = io.imread("tire.tif")
greyimg = rgb2gray(image)
#applying noise to image
noisy = random_noise(greyimg, mode='s&p', amount = 0.05)
#Filtering using built in function median
filtered = median(noisy) 
#Filtering using gaussian filter
gaus1 = gaussian(noisy,8)
gaus2 = gaussian(noisy,3) ## Best is around 1.2
gaus3 = gaussian(noisy,0.2)

array = []
array2 = []

array.append(greyimg)
array.append(noisy)
array.append(filtered)

array2.append(gaus1)
array2.append(gaus2)
array2.append(gaus3)

show_images(array)
show_images(array2)
```
